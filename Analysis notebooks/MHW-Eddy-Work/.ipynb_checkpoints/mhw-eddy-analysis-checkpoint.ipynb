# QMS515 - MHWs & Mesoscale Eddies

This notebook is going to explore the co-occurrence of Marine Heatwaves (MHWs) with mesoscale eddies in the EAC Extension region. The analysis will use remotely sensed sea surface temperature (SST) (NOAA-OISST) and sea surface height (SSH) (AVHRR-AVISO) data. 

The results from both the MHW and Eddy Tracking analyses have already been obtained, whereby `xmhw` and `py-eddy-tracker` were employed. 

This notebook will load in all the relevant data, and start with some exploratory interactive plots to get a general understanding of how the MHWs and eddies co-occur, or perhaps show that these are in fact independent of each other. Once a general understanding of this behaviour is characterised, I will then aim to obtain some sort of quantitative estimate of the likelihood of MHWs, given the present or future configuration of mesoscale eddies in the region.

```python
time_indices = np.unique(aviso_cc.time)
time_indices.shape
```

```python
def convert_day_to_date(days):
    base_date = datetime(1950,1,1)
    delta = timedelta(days=int(days))
    result_date = base_date + delta
    return result_date.strftime('%Y-%m-%d')

converted_dates = [convert_day_to_date(days) for days in aviso_ac.time]
converted_dates[-5:]
```

```python
def plot_eddies(i):
    # index = slice(aviso_cc.time[i-10], aviso_cc)
    mask = time_indices[i]
    ed = aviso_cc.extract_with_period(period = (time_indices[i], time_indices[i+5]))
    fig, ax = plt.subplots(figsize=(5,5))
    # ax=start_axes(f'{converted_dates[i]}')
    ax.set_xlim(aviso_ac['lon'].min(), aviso_ac['lon'].max())
    ax.set_ylim(aviso_ac['lat'].min(), aviso_ac['lat'].max())
    ed.display(ax)
```

```python
def dominant_freq(E):
    '''
    Computes dominant frequencies of column vectors
    of the input matrix using FFT.
    
    Args:
        E (numpy matrix): Matrix of eigenvectors (as columns)
    Returns:
        list of floats
    '''
    nfft = 2 ** 11                        # Sets the 'length of transformed axis' (essentially infinite?)
    fft = np.fft.fft(E, axis=0, n=nfft)   # Computes FFT - 
    fft = fft[0:nfft // 2]                # Obtains positive spectrum
    fft = np.abs(fft)**2                  # Amplitude squared (obtains power spectrum)
    freq = np.fft.fftfreq(nfft)           # range of frequencies of same length as nfft
    freq = freq[0:nfft // 2]              # Uniform power spectrum?
    freqs = freq[fft.argmax(axis=0)]      # Finds maximum single frequency from fft
    return freqs
```

```python
lat33_fft = dominant_freq(transect_lat33)
lat33_fft
```

```python
1/lat33_fft
```

```python
N = len(transect_lat33.time)
SR = 1 # samples per day

xf = np.fft.rfftfreq(N, 1/SR)
yf = np.fft.rfft(transect_lat33)
fig, ax = plt.subplots()
ax.plot(xf, np.abs(yf))
ax.set_xlim(0, 0.05)
ax.set_title(f'Frequency spectrum for SSH at {transect_lat33.latitude.values}, {transect_lat33.longitude.values}')
```

So, we have a dominant frequency around 0.01 cycles per day, which equates to ... one cycle per 100 days! That's pretty conclusive which is cool. So that's from the AVISO data that we see this dominant peak at 100 days. Would there be any signal in the MHW data?
